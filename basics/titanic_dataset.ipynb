```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic = sns.load_dataset("titanic")
print(titanic.head())

# Fill missing ages
titanic["age"].fillna(titanic["age"].mean(), inplace=True)

# Drop columns we won't use
titanic.drop(columns=["deck", "embark_town"], inplace=True)

# 1️⃣ Barplot: Survival by sex
sns.barplot(x="sex", y="survived", data=titanic)
plt.title("Survival Rate by Gender")
plt.show()

# 2️⃣ Histogram: Age distribution by survival
sns.histplot(data=titanic, x="age", hue="survived", multiple="stack")
plt.title("Age Distribution by Survival")
plt.show()

# 3️⃣ Heatmap: Correlations (numeric only).
# Select only numeric columns. numeric_titanic is a new dataset
numeric_titanic = titanic.select_dtypes(include="number")

# Plot heatmap
sns.heatmap(numeric_titanic.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
