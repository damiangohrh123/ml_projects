```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Initialize params
m, c = np.random.randn(1), np.random.randn(1)
lr = 0.01
epochs = 1000

# Gradient descent loop: update slope (m) and intercept (c) to fit the line
for _ in range(epochs):
    # Compute predicted values using current slope and intercept. (y-pred is an array)
    y_pred = m * X + c

    # Compute error (error is an array)
    error = y_pred - y

    # Gradient descent, update m and c based on learning rate
    m -= lr * (2/len(X)) * np.sum(error * X)
    c -= lr * (2/len(X)) * np.sum(error)

print("Our slope:", m, "Our intercept:", c)

# Compare with sklearn
model = LinearRegression()
model.fit(X, y)
print("Sklearn slope:", model.coef_[0], "Sklearn intercept:", model.intercept_)

# Plot
plt.scatter(X, y, alpha=0.5)
plt.plot(X, m*X + c, color="red", label="Our line")
plt.plot(X, model.predict(X), color="green", linestyle="--", label="Sklearn")
plt.legend()
plt.show()
```
